```python
num_samples_per_disease = 200  

expanded_data = []
for disease in diseases:
    for _ in range(num_samples_per_disease):
        entry = {symptom: 0 for symptom in symptoms}
        
        for symptom in symptoms:
            entry[symptom] = 1 if symptom in disease_symptoms[disease] or random.random() < 0.3 else 0
            
        for test in tests:
            entry[test] = disease_tests[disease][test]
        
        entry["disease"] = disease
        
        expanded_data.append(entry)

expanded_df = pd.DataFrame(expanded_data)
file_path = "expanded_dis_data.xlsx"
if not os.path.exists(file_path):
  expanded_df.to_excel("expanded_dis_data.xlsx", index=False, sheet_name="Data")

expanded_df
```

```python
from sklearn.model_selection import train_test_split

X = expanded_df.drop("disease", axis=1)  # Features: symptoms and tests
y = expanded_df["disease"]               # Target: disease

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svm_model = SVC(kernel='linear', random_state=42) 
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("SVM Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train logistic regression
lr_model = LogisticRegression(solver='liblinear', max_iter=1000, random_state=42)
lr_model.fit(X_train_scaled, y_train)
y_pred = lr_model.predict(X_test_scaled)
print("Logistic Regression Accuracy with Scaling:", accuracy_score(y_test, y_pred))
```

```python
feature_names = X_train.columns

user_input = input("Enter your symptoms and tests (comma-separated): ")
user_input_list = [item.strip().lower() for item in user_input.split(',')]


input_vector = {feature: 0 for feature in feature_names}

for feature in feature_names:
    if feature.lower() in user_input_list:
        input_vector[feature] = 1

input_df = pd.DataFrame([input_vector])
input_array = input_df.values

predicted_disease = lr_model.predict(input_array)[0]

# Output the prediction
print(f"The predicted disease is: {predicted_disease}")
```
